```python
from pyquery import PyQuery as pq
doc = pq(soup)
print(doc('h4'))
```

```python
house = []
for house in soup.select('.clearfix.text') :
        house.append({'地址':soup.select('.tit_shop')[0].text,'户型':soup.select('.tel_shop')[0].text})
```
